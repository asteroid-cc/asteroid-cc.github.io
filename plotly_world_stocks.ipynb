```python
def calculate_ema(price_data):
    # Calculate the simple moving average for the first 21 days
    try:
        sma = sum(price_data[:21]) / 21

        # Calculate the smoothing factor
        smoothing_factor = 2 / (21 + 1)

        # Calculate the initial EMA
        ema = (price_data[20] - sma) * smoothing_factor + sma

        # Calculate the EMA for the remaining days
        for i in range(21, len(price_data)):
            ema = (price_data[i] - ema) * smoothing_factor + ema
    except:
        ema = 0
    return ema
```
